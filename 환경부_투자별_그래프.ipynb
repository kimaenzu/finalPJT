```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings(action='ignore')
path = 'c:/finalPJT/csvData/'
```

```python
year = ['2012년', '2013년', '2014년', '2015년', '2016년', '2017년', '2018년', '2019년', '2020년']
```

```python
for y in year :
    df_d = pd.read_csv(path+"환경산업분류별(보호활동) 투자액.csv", thousands = ',', encoding='utf-8')
    df_d = df_d[df_d['분류.1'] == '소계'][2:]
    a = df_d[y]
    b = a.sort_values(ascending = False)
    index_list = list(b[:3].index)
    print(f"{y} TOP 3")
    for _ in index_list :
        print(df_d['분류'][_])
    print()
```

```python
top3 = ['자원관리관련서비스업','오염관리관련서비스업','오염관리관련건설업']
```

```python
top_1 = df_d[df_d['분류']== top3[0]]
top_2 = df_d[df_d['분류']== top3[1]]
top_3 = df_d[df_d['분류']== top3[2]]

df = pd.concat([top_1,top_2,top_3])
df = df.reset_index()
df
```

```python
#그래프 추출 범위지정
df_1=df.loc[0:5,['2016년', '2017년', '2018년', '2019년', '2020년']]
df_1=df_1.transpose()

#수출액, 환경부분수출액 2개의 막대그래프 합치기
cost0 = df_1[0][0:9]
cost1 = df_1[1][0:9]
cost1.index = ['2016년', '2017년', '2018년', '2019년', '2020년']

cost2 = df_1[2][0:9]
cost3 = df_1[3][0:9] 

cost4 = df_1[4][0:9]
cost5 = df_1[5][0:9] 

#바의 넓이, 그래프크기 지정
bar_width=0.25
fig, ax = plt.subplots(figsize=(12,6))

index = np.arange(len(cost1.index))
plt.bar(index,cost0,bar_width,label='export amount')
plt.bar(index,cost1,bar_width,label='environment')

plt.bar(index+bar_width,cost2,bar_width, label='export amount')
plt.bar(index+bar_width,cost3,bar_width, label='environment')

plt.bar(index+2*bar_width,cost4,bar_width, label='export amount')
plt.bar(index+2*bar_width,cost5,bar_width, label='environment')

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks(index,cost1.index)
#plt.yticks([0, 9])

plt.legend()
plt.xlabel('year')
plt.ylabel('money')  
```
